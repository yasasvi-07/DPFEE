```python
from matplotlib import pyplot as p
```

```python
import numpy as np
```

```python
x=np.array([10,20])
y=np.array([20,40])
```

```python
p.plot(x,y)
```

```python
p.plot(x,y,"o",)
```

```python
p.plot(x,y,marker="o")
```

```python
p.plot(x,y,marker="+")
```

```python
p.plot(x,y,marker="*")
```

```python
p.plot(x,y,marker="d")
```

```python
p.plot(x,y,'P-k')
```

```python
p.plot(x,y,'P:k')
```

```python
p.plot(x,y,'P--k')
```

```python
p.plot(x,y,'P-b')
```

```python
p.plot(x,y,'P--b',ms=10)
```

```python
p.plot(x,y,'p--b',ms=10)
```

```python
p.plot(x,y,'P--b',ms=10,mfc='k')
```

```python
p.plot(x,y,',--b',ms=10)
```

```python
p.plot(x,y,'>--b',ms=10)
```
